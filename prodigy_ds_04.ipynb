```python
import pandas as pd
import matplotlib.pyplot as plt
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')

social_media_data = {
    'Date': ['2024-05-01', '2024-05-02', '2024-05-03', '2024-05-04', '2024-05-05'],
    'Text': ['I love the new iPhone!', 'The customer service was terrible.', 'Excited about the new movie release!',
             'This product is amazing!', 'Disappointed with the quality.'],
    'Brand': ['Apple', 'Apple', 'MovieStudio', 'ProductX', 'ProductX']
}
df = pd.DataFrame(social_media_data)

sia = SentimentIntensityAnalyzer()

def get_sentiment_score(text):
    return sia.polarity_scores(text)['compound']

df['Sentiment'] = df['Text'].apply(get_sentiment_score)

plt.figure(figsize=(10, 6))
for brand, data in df.groupby('Brand'):
    plt.plot(data['Date'], data['Sentiment'], marker='o', label=brand)

plt.xlabel('Date')
plt.ylabel('Sentiment Score')
plt.title('Sentiment Analysis of Social Media Data')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
